##### 57. Income_Category의 카테고리를 map 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라
Unknown : N  
Less than $40K : a  
$40K - $60K : b  
$60K - $80K : c  
$80K - $120K : d  
$120K +’ : e  

```python
income_map = {
    "Unknown" : "N",
    "Less than $40K": "a",
    "$40K - $60K": "b",
    "$60K - $80K": "c",
    "$80K - $120K": "d",
    "$120K +": "e"
}

df["newIncome"] = df["Income_Category"].map(income_map)
# df['newIncome'] = df.Income_Category.map(lambda x: dic[x])
df.head()
```

```python
def split(money):
    if money == "Unknown": return "N"
    elif money == "Less than $40K": return "a"
    elif money == "$40K - $60K": return "b"
    elif money == "$60K - $80K": return "c"
    elif money == "$80K - $120K": return "d"
    elif money == "$120K +": return "e"

df["newIncome"] = df["Income_Category"].map(split)
df.head()
```

##### 58. Income_Category의 카테고리를 apply 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라 Unknown : N
Less than $40K : a  
$40K - $60K : b  
$60K - $80K : c  
$80K - $120K : d  
$120K +’ : e

```python
def changeCategory(money):
    if money == "Unknown": return "N"
    elif money == "Less than $40K": return "a"
    elif money == "$40K - $60K": return "b"
    elif money == "$60K - $80K": return "c"
    elif money == "$80K - $120K": return "d"
    elif money == "$120K +": return "e"

df["newIncome"] = df["Income_Category"].apply(changeCategory)
df.head()
```

##### 59. Customer_Age의 값을 이용하여 나이 구간을 AgeState 컬럼으로 정의하라. (0~9 : 0 , 10~19 :10 , 20~29 :20 … 각 구간의 빈도수를 출력하라

```python
df['AgeState']  = df["Customer_Age"].map(lambda x: x // 10 * 10)
df["AgeState"].value_counts().sort_index()
```

##### 60. Education_Level의 값중 Graduate단어가 포함되는 값은 1 그렇지 않은 경우에는 0으로 변경하여 newEduLevel 컬럼을 정의하고 빈도수를 출력하라

```python
df["newEduLevel"] = df["Education_Level"].map(lambda x: 1 if "Graduate" in x else 0)
df["newEduLevel"].value_counts().sort_index()
```

##### 60. Credit_Limit 컬럼값이 4500 이상인 경우 1 그외의 경우에는 모두 0으로 하는 newLimit 정의하라. newLimit 각 값들의 빈도수를 출력하라

```python
df["newLimit"] = df['Credit_Limit'].map(lambda x: 1 if x >= 4500 else 0)
df["newLimit"].value_counts().sort_index()
```

##### 62. Marital_Status 컬럼값이 Married 이고 Card_Category 컬럼의 값이 Platinum인 경우 1 그외의 경우에는 모두 0으로 하는 newState컬럼을 정의하라. newState의 각 값들의 빈도수를 출력하라

```python
def check(x):
    if x.Marital_Status == "Married" and x.Card_Category == "Platinum": return 1
    else: return 0

df["newState"] = df.apply(check, axis = 1) # axis = 1 : 행 단위 으로 함수 적용 => 각 행을 Series로 전달해 함수 실행
df["newState"].value_counts()
```

##### 63. Gender 컬럼값 M인 경우 male F인 경우 female로 값을 변경하여 Gender 컬럼에 새롭게 정의하라. 각 value의 빈도를 출력하라

```python
df["Gender"] = df["Gender"].map(lambda x: "male" if x == "M" else "female")
df["Gender"].value_counts()
```

##### 67. Yr_Mo_Dy에 년도가 2061년 이상의 경우에는 모두 잘못된 데이터이다. 해당경우의 값은 100을 빼서 새롭게 날짜를 Yr_Mo_Dy 컬럼에 정의하라

```python
print(df.head())

def yearCheck(x):
    import datetime

    year = x.year - 100 if x.year >= 2061 else x.year
    return pd.to_datetime(datetime.date(year, x.month, x.day))

df['Yr_Mo_Dy'] = df['Yr_Mo_Dy'].apply(yearCheck)
df.head()
```

##### 76. 년-월-일:시 컬럼을 pandas에서 인식할 수 있는 datetime 형태로 변경하라. 서울시의 제공데이터의 경우 0시가 24시로 표현된다
서울시 미세먼지 데이터 : https://www.airkorea.or.kr/web/realSearch?pMENU_NO=97  
DataUrl = ‘https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv’

```python
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv")
df.head()
```

```python
def change(x):
    import datetime
    date = x.split(":")[0]
    hour = x.split(":")[1]

    if hour == "24":
        hour = "00:00:00"
        return pd.to_datetime(date + " " + hour) + datetime.timedelta(days = 1)
    else:
        hour = hour + ":00:00"
        return pd.to_datetime(date + " " + hour)
    
df["(년-월-일:시)"] = df["(년-월-일:시)"].apply(change)
df.head()
```
