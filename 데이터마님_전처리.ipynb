##### 58. Income_Category의 카테고리를 apply 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라

Less than $40K : a

```python
def category(x):
    if x == "Unknown": return "N"
    elif x == "Less than $40K" : return "a"
    elif x == "$40K - $60K" : return "b"
    elif x == "$60K - $80K" : return "c"
    elif x == "$80K - $120K" : return "d"
    elif x == "$120K +" : return "e"

df["newIncome"] = df["Income_Category"].apply(category)
```

##### 59. Customer_Age의 값을 이용하여 나이 구간을 AgeState 컬럼으로 정의하라. (0~9 : 0 , 10~19 :10 , 20~29 :20 … 각 구간의 빈도수를 출력하라

##### 62. ❌ Marital_Status 컬럼값이 Married 이고 Card_Category 컬럼의 값이 Platinum인 경우 1 그외의 경우에는 모두 0으로 하는 newState컬럼을 정의하라. newState의 각 값들의 빈도수를 출력하라

##### 76. ❌ 년-월-일:시 컬럼을 pandas에서 인식할 수 있는 datetime 형태로 변경하라. 서울시의 제공데이터의 경우 0시가 24시로 표현된다
서울시 미세먼지 데이터 : https://www.airkorea.or.kr/web/realSearch?pMENU_NO=97  
DataUrl = ‘https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv’

```python
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv")
```

```python
import datetime as dt

def change(x):
    date = x.split(":")[0]
    hour = x.split(":")[1]

    if hour == "24":
        hour = "00:00:00"
        return pd.to_datetime(date + " " + hour) + dt.timedelta(days = 1)
    else:
        hour = hour + ":00:00"
        return pd.to_datetime(date + " " + hour)
    
df["(년-월-일:시)"] = df["(년-월-일:시)"].apply(change)
df.head()
```

##### 77. 일자별 영어 요일 이름을 dayName 컬럼에 저장하라

```python
df["dayName"] = df["(년-월-일:시)"].dt.day_name()
df
```

##### 78. 일자별 각 PM10등급의 빈도수를 파악하라

```python
ans = df.groupby(["dayName", "PM10등급"],as_index=False).size()
ans.pivot_table(index = "dayName", columns = "PM10등급", values = "size")
```

##### 92. df3과 df4 데이터를 하나의 데이터 프레임으로 합쳐라. 둘다 포함하고 있는 년도에 대해서만 고려한다

```python
ans = pd.concat([df3, df4], join = "outer").fillna(0)
ans
```
