#### 문제: 다음은 22명의 학생들이 국어시험에서 받은 점수이다. 학생들의 평균이 75보다 크다고 할 수 있는가?
귀무가설(H0): 모평균은 mu와 같다. (μ = mu), 학생들의 평균은 75이다  
대립가설(H1): 모평균은 mu보다 크다. (μ > mu), 학생들의 평균은 75보다 크다

```python
# 검정통계량, p-value, 검정결과를 출력하시오

# 데이터
scores = [75, 80, 68, 72, 77, 82, 81, 79, 70, 74, 76, 78, 81, 73, 81, 78, 75, 72, 74, 79, 78, 79]

from scipy import stats
mu = 75
alpha = 0.05

statistic, p_value = stats.ttest_1samp(scores, mu, alternative = "greater")
print("검정통계량 :", statistic)
print("p-value :", p_value)

if p_value < alpha: print("H0 기각")
else: print("H1 채택")
```

#### 어떤 특정 약물을 복용한 사람들의 평균 체온이 복용하지 않은 사람들의 평균 체온과 유의미하게 다른지 검정해보려고 합니다.

가정: 
- 약물을 복용한 그룹과 복용하지 않은 그룹의 체온 데이터가 각각 주어져 있다고 가정합니다.
- 각 그룹의 체온은 정규분포를 따른다고 가정합니다.

```python
# 가설 설정
# H0: 약물을 복용한 그룹과 복용하지 않은 그룹의 평균 체온은 유의미한 차이가 없다.
# H1: 약물을 복용한 그룹과 복용하지 않은 그룹의 평균 체온은 유의미한 차이가 있다.

# 데이터 수집
group1 = [36.8, 36.7, 37.1, 36.9, 37.2, 36.8, 36.9, 37.1, 36.7, 37.1]
group2 = [36.5, 36.6, 36.3, 36.6, 36.9, 36.7, 36.7, 36.8, 36.5, 36.7]
alpha = 0.05

statistic, p_value = stats.ttest_ind(group1, group2)
print("검정통계량 :", statistic)
print("p-value :", p_value)

if p_value < alpha: print("H0 기각")
else: print("H1 채택")
```

고객 정보를 나타낸 데이터이다. 주어진 데이터에서 500개 중 앞에서부터 350개는 train으로, 150개는 test 데이터로 나눈다. 모델을 학습(적합)할 때는 train 데이터를 사용하고, 예측할 때는 test 데이터를 사용한다. 모델은 로지스틱 회귀를 써서 고객이 특정 제품을 구매할지 여부를 예측하되, 페널티는 부과하지 않는다.  
  
종속변수: purchase (0: 구매 안 함, 1: 구매 함)
  
1. income 변수를 독립변수로 purchase를 종속변수로 사용하여 로지스틱 회귀 모형을 만들고, income 변수가 한 단위 증가할 때 구매할 오즈비 값을 계산하시오. (반올림하여 소수 넷째자리까지 계산)
  
2. 독립변수 income만 사용해 학습한 모델에서 test 데이터의 purchase를 예측하고, accuracy (정확도)를 구하시오. (반올림하여 소수 셋째자리까지 계산)
  
3. 독립변수 income만 사용해 학습한 모델의 로짓 우도를 계산하시오.
  
4. 독립변수 income만 사용해 학습한 모델의 유의확률(p-value)를 구하시오.

```python
import pandas as pd
import numpy as np
df = pd.read_csv("../data/Customer_Data.csv")

df.head()
```

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
from sklearn.metrics import accuracy_score

df = pd.read_csv("../data/Customer_Data.csv")

train = df[:350]
test = df[350:]  # 수정: 350~500

model1 = smf.logit("purchase ~ income", data=train).fit()

# 1. 오즈비
odds_ratio = np.exp(model1.params["income"])
print("1. 오즈비 :", round(odds_ratio, 4))

# 2. 정확도
pred_prob = model1.predict(test)
pred_class = (pred_prob >= 0.5).astype(int)
acc = accuracy_score(test["purchase"], pred_class)
print("2. 정확도 :", round(acc, 3))

# 3. 로그 우도 (log-likelihood)
print("3. 로그 우도 :", round(model1.llf, 4))

# 4. p-value (income 변수)
print("4. p-value :", round(model1.pvalues["income"], 4))
```
